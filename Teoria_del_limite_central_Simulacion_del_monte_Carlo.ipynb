```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Configuración de la simulación
n_experimentos = 5000
n_dados = 50

# Generamos la matriz de lanzamientos (5000 experimentos, cada uno con 50 dados)
# Cada valor es un entero entre 1 y 6
lanzamientos = np.random.randint(1, 7, size=(n_experimentos, n_dados))

# 2. Calculamos las medias de cada experimento
medias = np.mean(lanzamientos, axis=1)

# 3. Mostrar resultados estadísticos detallados
print("--- Resumen de la Simulación Monte Carlo (TLC) ---")
print(f"Total de experimentos: {n_experimentos}")
print(f"Dados lanzados por experimento: {n_dados}")
print(f"Media global de todos los resultados: {np.mean(medias):.4f}")
print(f"Desviación estándar de las medias: {np.std(medias):.4f}")
print(f"Valor mínimo obtenido: {np.min(medias):.4f}")
print(f"Valor máximo obtenido: {np.max(medias):.4f}")
print("-" * 48)

# 4. Visualización de la convergencia a la Distribución Normal
plt.figure(figsize=(10, 6))
plt.hist(medias, bins=30, color='green', edgecolor='black', density=True, alpha=0.7)
plt.axvline(np.mean(medias), color='red', linestyle='dashed', linewidth=2, label=f'Media: {np.mean(medias):.2f}')
plt.title("Distribución de Medias (Teorema del Límite Central)")
plt.xlabel("Promedio de los 50 lanzamientos")
plt.ylabel("Frecuencia (Densidad)")
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```
